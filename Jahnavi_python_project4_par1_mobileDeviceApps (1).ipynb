# Question 5
5. Check the unique categories of the column 'Category', Is there any invalid category? If yes, drop them

```python
# checking and printing the unique categories

unique_categories = data['Category'].unique
print("The unique categories are \n", unique_categories,"\n\n")

# using boolean indexing to filter out invalid categories from the dataset.
invalid_categories = ['invalid_category1', 'invalid_category2']
data = data[~data['Category'].isin(invalid_categories)]

# printing the unique categories after dropping the invalid category
unique_categories = data['Category'].unique()
print("Unique Categories after dropping invalid categories:\n", unique_categories)
```

# Question 6

```python
# checking the number of missing values using 'isnull' function

print("The number of missing values in the rating column are:",data['Rating'].isnull().sum())
data.dropna(subset = ['Rating'], inplace = True)

# create a new column as 'Rating_category' by converting ratings to high and low categories(>3.5 is high rest low)

data['Rating_category'] = ['high' if rating > 3.5 else 'low' for rating in data['Rating']]
```

# Question 8

```python
#  Convert the column "Reviews" to numeric data type
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')

# checking the presence of outliers in the column
sns.boxplot(data['Reviews'])
plt.xlabel('Reviews')
plt.show()

# handling the outliers using log transformation approach
data['Reviews'] = data['Reviews'].apply(lambda x: np.log1p(x))
```

# Question 9

```python
# Converting the column 'Size' with alphanumeric to numbers and converting into suitable data type

data['Size'] = data['Size'].replace('M','1000000')
data['Size'] = data['Size'].replace('K','1000')
data['Size'] = data['Size'].replace('Varies with device','1000000')

# The errors='coerce' parameter ensures that any non-numeric values are converted to NaN.
data['Size'] = pd.to_numeric(data['Size'], errors='coerce')
data = data.dropna(subset=['Size'])
```

# Question 10

```python
# Check the column 'Installs' and treating the unwanted characters and convert the column into a suitable data type.

# The str.replace() function is used to replace the unwanted characters with an empty string.
data['Installs'] = data['Installs'].str.replace(',', '')
data['Installs'] = data['Installs'].str.replace('+', '')
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')
data['Installs'].dtype
```

```python
data['Installs'].dtype
```

# Question 11

```python
# Check the column 'Price' , remove the unwanted characters and convert the column into a suitable data type.

data['Price'] = data['Price'].str.replace('$','')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
```
